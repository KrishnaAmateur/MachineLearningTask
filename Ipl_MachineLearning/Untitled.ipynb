```python
import pandas as pd
import sklearn
```

```python
df=pd.read_csv('ipl_data.csv')
```

```python
df.head()
```

```python
df.drop(['mid','batsman','bowler','striker','non-striker'],axis=1,inplace=True)
```

```python
df['bat_team'].unique()
```

```python
df['venue'].unique()
```

```python
df['venue'] = df['venue'].replace({'Punjab Cricket Association Stadium, Mohali':'Punjab Cricket Association IS Bindra Stadium, Mohali'})
```

```python
df['venue'].unique()
```

```python
selected_venues = ['M Chinnaswamy Stadium',
       'Punjab Cricket Association IS Bindra Stadium, Mohali',
       'Feroz Shah Kotla', 'Wankhede Stadium', 'Eden Gardens',
       'Sawai Mansingh Stadium',
       'Rajiv Gandhi International Stadium, Uppal',
       'MA Chidambaram Stadium, Chepauk','Dr. Y.S. Rajasekhara Reddy ACA-VDCA Cricket Stadium','Sheikh Zayed Stadium',
       'Sharjah Cricket Stadium', 'Dubai International Cricket Stadium']
```

```python
current_teams=['Kolkata Knight Riders', 'Chennai Super Kings', 'Rajasthan Royals',
       'Mumbai Indians','Kings XI Punjab',
       'Royal Challengers Bangalore', 'Delhi Daredevils',
       'Sunrisers Hyderabad']
```

```python
df=df[(df['bat_team'].isin(current_teams)) & (df['bowl_team'].isin(current_teams)) & (df['venue'].isin(selected_venues)) ]
```

```python
df['venue'].unique()
```

```python
df['bat_team'].unique()
```

```python
df.shape
```

```python
df=df[df['overs']>=5.0]
```

```python
df.head()
```

```python
df.shape
```

```python
df.info()
```

```python
df['date']= pd.to_datetime(df['date'])
```

```python
df.info()
```

```python
df['year']=df['date'].dt.year
```

```python
df.head()
```

```python
df.drop('date',axis=1,inplace=True)
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
one_hot_encode = pd.get_dummies(df,['venue','bat_team','bowl_team'])
```

```python
one_hot_encode.head()
```

```python
one_hot_encode.columns
```

```python
one_hot_encode=one_hot_encode[['bat_team_Chennai Super Kings', 'bat_team_Delhi Daredevils',
       'bat_team_Kings XI Punjab', 'bat_team_Kolkata Knight Riders',
       'bat_team_Mumbai Indians', 'bat_team_Rajasthan Royals',
       'bat_team_Royal Challengers Bangalore', 'bat_team_Sunrisers Hyderabad',
       'bowl_team_Chennai Super Kings', 'bowl_team_Delhi Daredevils',
       'bowl_team_Kings XI Punjab', 'bowl_team_Kolkata Knight Riders',
       'bowl_team_Mumbai Indians', 'bowl_team_Rajasthan Royals',
       'bowl_team_Royal Challengers Bangalore',
       'bowl_team_Sunrisers Hyderabad','venue_Dr. Y.S. Rajasekhara Reddy ACA-VDCA Cricket Stadium',
       'venue_Dubai International Cricket Stadium', 'venue_Eden Gardens',
       'venue_Feroz Shah Kotla', 'venue_M Chinnaswamy Stadium',
       'venue_MA Chidambaram Stadium, Chepauk',
       'venue_Punjab Cricket Association IS Bindra Stadium, Mohali',
       'venue_Rajiv Gandhi International Stadium, Uppal',
       'venue_Sawai Mansingh Stadium', 'venue_Sharjah Cricket Stadium',
       'venue_Sheikh Zayed Stadium', 'venue_Wankhede Stadium','runs', 'wickets', 'overs', 'runs_last_5', 'wickets_last_5','year','total']]
```

```python
one_hot_encode.head()
```

```python
X_train = one_hot_encode.drop('total',axis=1)[one_hot_encode['year']<=2016]
X_test = one_hot_encode.drop('total',axis=1)[one_hot_encode['year']>=2017]
```

```python
y_train = one_hot_encode['total'][one_hot_encode['year']<=2016]
y_test = one_hot_encode['total'][one_hot_encode['year']>=2017]
```

```python
X_train.drop('year',axis=1,inplace=True)
X_test.drop('year',axis=1,inplace=True)
```

```python
from sklearn.linear_model import LinearRegression
regression = LinearRegression()
regression.fit(X_train,y_train)
```

```python
y_predict=regression.predict(X_test)
y_predict
```

```python
import pickle
filename = 'score.pkl'
pickle.dump(regression,open(filename,'wb'))
```
